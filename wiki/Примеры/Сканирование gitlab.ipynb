```python
import requests as r
import pandas as pd
```

```python
# maximum 100 per page
# see https://docs.gitlab.com/ee/api/index.html#pagination
projects = r.get(r'http://gitlab.lis/api/v4/projects',{'per_page':100,'page':1}).json()
len(projects),[x['name'] for x in projects]
```

```python
projects[0]
```

```python
def fill_project_description(project):
    x=dict(project)
    if not x['description'] and x['readme_url']:
        x['description'] = r.get(x['readme_url'].replace('/blob/','/raw/')).text[:200]
    return x
```

```python
df = pd.DataFrame([fill_project_description(x) for x in projects])
df
```

```python
df[['description','name','web_url']].to_excel('Проекты gitlab.xlsx',index=False)
```
